```python
import json

a_json = open('../result/naver_news/news_한동훈_naver_20220620_20220623__202206.json', 'r', encoding = 'utf-8')
a_dict = json.load(a_json)
```

```python
a_dict
```

```python
import re

def remove_emoji(comment):
    emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"
            u"\U0001F300-\U0001F5FF"
            u"\U0001F680-\U0001F6FF"
            u"\U0001F1E0-\U0001F1FF"
            "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', comment)

for date in a_dict.keys():
    for url in a_dict[date].keys():
        for comment in a_dict[date][url]["comments"]:
            print(date, remove_emoji(comment))
```

```python
import pandas as pd
from openpyxl import Workbook

wb = Workbook()
ws = wb.active
ws.title = "crawling_data"

i = 1
for date in a_dict.keys():
    for url in a_dict[date].keys():
        for comment in a_dict[date][url]["comments"]:
            ws.cell(i, 1, date)
            ws.cell(i, 2, remove_emoji(comment))
            i+=1

wb.save('crawling_data.xlsx')
```
